```python
import torch

# 1. 张量基础
print(f"PyTorch 版本: {torch.__version__}")

# 标量（0维张量）
scale = torch.tensor(7)
print(f"标量: {scale}, ndim: {scale.ndim}")

# 向量（1维张量）
vector = torch.tensor([7, 7, 8], dtype=torch.float64)
print(f"向量: {vector}, shape: {vector.shape}, dtype: {vector.dtype}")

# 随机张量
random_tensor = torch.rand(size=(2, 3))
print(f"随机张量:\n{random_tensor}, dtype: {random_tensor.dtype}")
```

```python
# 2. MPS 设备（Mac GPU 加速）
some_tensor = torch.rand(3, 4, device="mps")

print(f"张量:\n{some_tensor}")
print(f"Shape: {some_tensor.shape}")
print(f"Device: {some_tensor.device}")
```

```python
# 3. 基础运算
tensor = torch.tensor([1, 2, 3])
print(f"加法: {tensor + 10}")
print(f"乘法: {tensor * 10}")
print(f"元素积: {tensor * tensor}")
```

```python
# 4. 矩阵乘法
# 两种写法等价：torch.matmul(a, b) 和 a @ b
print(f"点积: {torch.matmul(tensor, tensor)}")
print(f"@ 运算符: {tensor @ tensor}")

# 2x3 @ 3x2 -> 2x2
matrix = torch.tensor([[7, 8, 9], [10, 11, 12]], dtype=torch.float32)
print(f"矩阵乘法:\n{matrix @ matrix.T}")
```

```python
tensor @ tensor  # 矩阵乘法简写
```

```python
# 2x3 矩阵乘转置 -> 2x2
tensor = torch.tensor([[7, 8, 9], [10, 11, 12]], dtype=torch.float32)
print(f"矩阵乘转置:\n{tensor @ tensor.T}")
```

```python
# 5. 线性层（神经网络基础）
tensor_A = torch.tensor([[1, 2], [3, 4], [5, 6]], dtype=torch.float32)
torch.manual_seed(42)

# Linear(in_features=2, out_features=10) -> 2维输入，10维输出
linear = torch.nn.Linear(in_features=2, out_features=10)
output = linear(tensor_A)

print(f"输入: {tensor_A.shape} -> 输出: {output.shape}")
print(f"权重: {linear.weight.shape}")
```
